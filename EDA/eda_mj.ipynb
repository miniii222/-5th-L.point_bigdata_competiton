### K-means with word2vec - with product

```python
np.array(kwd_merge['CLAC1_NM']+kwd_merge['KWD_NM'])
```

```python
# -*- coding: utf-8 -*-
from gensim.models import Word2Vec
from nltk.cluster import KMeansClusterer
import nltk

from sklearn import cluster
from sklearn import metrics
 
# training data
kwdlist=np.array(kwd_merge['CLAC1_NM']+kwd_merge['KWD_NM'])
 
# training model
model = Word2Vec(kwdlist, min_count=1)
 
# get vector data
X = model[model.wv.vocab]
print (X)
```

```python
#master['CLAC1_NM'].value_counts()
```

```python
print (model.similarity('퓨마', '스포츠패션'))
print (model.similarity('여성', '아동'))
print (model.similarity('화장품', '생활/주방가전'))
print (model.similarity('아동', '유아동의류'))
print (model.similarity('아동', '생활/주방가전'))
print (model.similarity('구두', '패션잡화'))
print (model.similarity('퓨마', '아웃도어/레저'))
print (model.similarity('화장품', '리리코스'))
```

```python
NUM_CLUSTERS=3
kclusterer = KMeansClusterer(NUM_CLUSTERS, distance=nltk.cluster.util.cosine_distance, repeats=25)
assigned_clusters = kclusterer.cluster(X, assign_clusters=True)
print (assigned_clusters)
 
#words = list(model.wv.vocab)
#for i, word in enumerate(words):  
#    print (word + ":" + str(assigned_clusters[i]))
 

kmeans = cluster.KMeans(n_clusters=NUM_CLUSTERS)
kmeans_fit=kmeans.fit(X)
 
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
 
print ("Cluster id labels for inputted data")
print (labels)
print ("Centroids data")
print (centroids)
 
print ("Score (Opposite of the value of X on the K-means objective which is Sum of distances of samples to their closest cluster center):")
print (kmeans.score(X))
```

```python
words = list(model.wv.vocab)
for i, word in enumerate(words):  
    print (word + ":" + str(assigned_clusters[i]))
```
